```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
import os
import warnings
warnings.filterwarnings("ignore")

from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import (
    train_test_split, StratifiedKFold,
    GridSearchCV, cross_val_score
)
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve,
    mean_absolute_error, mean_squared_error, r2_score
)
from sklearn.utils.class_weight import compute_sample_weight

MODELS_DIR = r"D:\Programs\supply_chain_ai\backend\app\ml"
DATA_DIR   = r"D:\Programs\supply_chain_ai\data"
```

```python
df = pd.read_csv(os.path.join(DATA_DIR, "processed_dataset.csv"))

print(f"Shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nNull check:")
print(df.isnull().sum())
print(f"\nClass balance:")
print(df["late_delivery_risk"].value_counts(normalize=True).round(3))
print(f"\nSample:")
print(df.head(3))
```

```python
mode_lookup = pd.read_csv(os.path.join(MODELS_DIR, "mode_delay_lookup.csv"))
print("Mode delay lookup :")
print(mode_lookup)

encoded_to_rate = {
    0: mode_lookup[mode_lookup["shipping_mode"]=="Standard Class"]["mode_delay_rate"].values[0],
    1: mode_lookup[mode_lookup["shipping_mode"]=="Second Class"]["mode_delay_rate"].values[0],
    2: mode_lookup[mode_lookup["shipping_mode"]=="First Class"]["mode_delay_rate"].values[0],
    3: mode_lookup[mode_lookup["shipping_mode"]=="Same Day"]["mode_delay_rate"].values[0],
}

df["mode_delay_rate"] = df["shipping_mode_encoded"].map(encoded_to_rate)

print(f"\nmode_delay_rate stats:")
print(df["mode_delay_rate"].describe().round(3))
print(f"\nDelay rate by mode_delay_rate (validation):")
print(df.groupby("mode_delay_rate")["late_delivery_risk"].mean().round(3))
```

## Feature Matrix

```python
FEATURE_COLS = [
    "days_shipping_scheduled",  # base journey time expectation
    "shipping_mode_encoded",    # integer for tree splits
    "mode_delay_rate",          # historical delay rate per mode — new signal
    "zone_id_origin",           # K-Means cluster from Phase 2
    "zone_risk_score",          # zone average delay rate from Phase 2
    "anomaly_flag",             # Isolation Forest flag from Phase 2
]

TARGET_CLASS = "late_delivery_risk"
TARGET_REG   = "delay_minutes"

df_model = df[FEATURE_COLS + [TARGET_CLASS, TARGET_REG]].dropna().copy()

print(f"Features ({len(FEATURE_COLS)}): {FEATURE_COLS}")
print(f"\nDataset shape: {df_model.shape}")
print(f"\nNull check — must all be 0:")
print(df_model.isnull().sum())
print(f"\nFeature stats:")
print(df_model[FEATURE_COLS].describe().round(3))
```

## Train-Test-Split

```python
X = df_model[FEATURE_COLS].copy()
y_class = df_model[TARGET_CLASS].copy()
y_reg = df_model[TARGET_REG].copy()

X_train, X_test, y_train_c, y_test_c, y_train_r, y_test_r = train_test_split(
    X, y_class, y_reg,
    test_size=0.2,
    random_state=42,
    stratify=y_class
)

sample_weights = compute_sample_weight(class_weight="balanced", y=y_train_c)

print(f"Train: {X_train.shape[0]} rows")
print(f"Test:  {X_test.shape[0]} rows")
print(f"Train balance: {y_train_c.value_counts(normalize=True).round(3).to_dict()}")
```

## XGBoost Classifier - With the best hyperparameter (tested by GridSearchCV)

```python
# n_estimators=300, max_depth=6, learning_rate=0.1
# subsample=0.8, colsample_bytree=0.9, gamma=0, min_child_weight=1
clf = XGBClassifier(
    n_estimators = 300,
    max_depth=6,
    learning_rate = 0.05,
    subsample=0.8,
    colsample_bytree = 0.9,
    random_state = 42,
    eval_metric = "logloss"
)
clf.fit(X_train,y_train_c,sample_weight = sample_weights)
```

```python
y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:,1]
auc = roc_auc_score(y_test_c,y_prob)

print(f"Test AUC-ROC: {auc:.4f}")
print(f"\nClassification report:")
print(classification_report(y_test_c, y_pred))
```

## Overfit Check

```python
train_prob = clf.predict_proba(X_train)[:, 1]
train_auc = roc_auc_score(y_train_c, train_prob)
test_auc = auc
gap = train_auc - test_auc

print(f"Train AUC : {train_auc:.4f}")
print(f"Test AUC : {test_auc:.4f}")
print(f"Gap : {gap:.4f}")
```

## ROC Curve & Confusion Matrix

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fpr, tpr, _ = roc_curve(y_test_c, y_prob)
axes[0].plot(fpr, tpr, color="steelblue", lw=2, label=f"AUC = {test_auc:.3f}")
axes[0].plot([0, 1], [0, 1], "k--", lw=1)
axes[0].fill_between(fpr, tpr, alpha=0.1, color="steelblue")
axes[0].set_title("ROC Curve — Delay Risk Classifier")
axes[0].set_xlabel("False Positive Rate")
axes[0].set_ylabel("True Positive Rate")
axes[0].legend()

cm = confusion_matrix(y_test_c, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=axes[1],
            xticklabels=["On Time", "Delayed"],
            yticklabels=["On Time", "Delayed"])
axes[1].set_title("Confusion Matrix")
axes[1].set_ylabel("Actual")
axes[1].set_xlabel("Predicted")

plt.tight_layout()
plt.show()

tn, fp, fn, tp = cm.ravel()
print(f"Missed real delays (False Negatives): {fn}  ← most costly in supply chain")
print(f"False alarms (False Positives): {fp}")
```

## Feature Importance

```python
importance_df = pd.DataFrame({
    "feature":    FEATURE_COLS,
    "importance": clf.feature_importances_
}).sort_values("importance", ascending=True)

plt.figure(figsize=(9, 5))
plt.barh(importance_df["feature"], importance_df["importance"], color="steelblue")
plt.title("Feature Importance — XGBoost Classifier")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()

print("Feature importance ranking:")
print(importance_df.sort_values("importance", ascending=False).to_string(index=False))
```

## XGBoost Regressor

```python
mask_train = y_train_r > 0
mask_test  = y_test_r > 0

print(f"Training regressor on {mask_train.sum()} delayed shipments")
print(f"Testing  regressor on {mask_test.sum()} delayed shipments")
```

```python
reg = XGBRegressor(
    n_estimators = 300,
    max_depth = 6,
    learning_rate = 0.1,
    subsample = 0.8,
    colsample_bytree = 0.9,
    min_child_weight = 1,
    gamma = 0,
    random_state = 42,
    verbosity = 0
)
reg.fit(X_train[mask_train], y_train_r[mask_train])
```

```python
y_pred_reg = reg.predict(X_test[mask_test])
train_pred = reg.predict(X_train[mask_train])
```

```python
mae       = mean_absolute_error(y_test_r[mask_test], y_pred_reg)
rmse      = np.sqrt(mean_squared_error(y_test_r[mask_test], y_pred_reg))
r2        = r2_score(y_test_r[mask_test], y_pred_reg)
train_mae = mean_absolute_error(y_train_r[mask_train], train_pred)

print(f"\nRegressor results:")
print(f"Train MAE : {train_mae:.0f} min  ({train_mae/60:.1f} hrs)")
print(f"Test MAE : {mae:.0f} min  ({mae/60:.1f} hrs)")
print(f"Gap : {mae - train_mae:.0f} min")
print(f"RMSE : {rmse:.0f} min")
print(f"R² : {r2:.4f}")
```

```python
joblib.dump(clf, os.path.join(MODELS_DIR, "xgb_classifier.joblib"))
joblib.dump(reg, os.path.join(MODELS_DIR, "xgb_regressor.joblib"))

with open(os.path.join(MODELS_DIR, "feature_cols.json"), "w") as f:
    json.dump(FEATURE_COLS, f, indent=2)

with open(os.path.join(MODELS_DIR, "mode_rate_map.json"), "w") as f:
    json.dump({str(k): v for k, v in encoded_to_rate.items()}, f, indent=2)

print("Saved:")
for fname in sorted(os.listdir(MODELS_DIR)):
    size = os.path.getsize(os.path.join(MODELS_DIR, fname))
    print(f"{fname:45s}  {size/1024:6.1f} KB")

print(f"\nClassifier AUC-ROC : {test_auc:.4f}")
print(f"Train/Test gap : {gap:.4f}")
print(f"Regressor MAE : {mae:.0f} min ({mae/60:.1f} hrs)")
print(f"Regressor R² : {r2:.4f}")
print(f"Features : {FEATURE_COLS}")
```
